```python
import sqlalchemy as db

#Establishing connection with the database.
engine = db.create_engine('mysql://root:2000@localhost/Organization')

connection = engine.connect()
metadata = db.MetaData()
```

```python
employee = db.Table('employee', metadata, autoload=True, autoload_with=engine)
```

```python
import pandas as pd

query = db.select([employee]) 
ResultProxy = connection.execute(query)
Result = ResultProxy.fetchall()
df = pd.DataFrame(Result, columns = Result[0].keys())
```

```python
df
```

```python
query = db.select([employee.columns.empname, employee.columns.salary]).where(employee.columns.salary > 5000)
resultproxy2 = connection.execute(query)
result2 = resultproxy2.fetchall()
df2 = pd.DataFrame(result2, columns = result2[0].keys())
df2
```

```python
query = db.select([employee.columns.empname.distinct()])
resultproxy3 = connection.execute(query)
result3 = resultproxy3.fetchall()
df3 = pd.DataFrame(result3, columns = ["Name"])
df3
```

```python
query = db.select([employee.columns.empname, employee.columns.salary]).where(employee.columns.salary.in_([2500, 6500, 10000]))
resultproxy4 = connection.execute(query)
result4 = resultproxy4.fetchall()
df4 = pd.DataFrame(result4, columns = result4[0].keys())
df4
```

```python
query = db.select([employee.columns.departmentid, db.func.count(employee.columns.empname).label("count")]).group_by(employee.columns.departmentid)
resultproxy5 = connection.execute(query)
result5 = resultproxy5.fetchall()
df5 = pd.DataFrame(result5, columns = result5[0].keys())
df5
```

```python
query = db.select([employee]).order_by(employee.columns.salary)
resultproxy6 = connection.execute(query)
result6 = resultproxy6.fetchall()
df6 = pd.DataFrame(result6, columns = result6[0].keys())
df6
```
